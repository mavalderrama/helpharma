```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import re
```

```python
import unicodedata
def rem_accents(text):
    return unicodedata.normalize('NFD', text)\
           .encode('ascii', 'ignore')\
           .decode("utf-8")
```

```python
consulta = pd.read_excel("data/Consulta.xlsx")
consulta.dropna(subset=["ID"],inplace=True)
consulta["ID"] = consulta["ID"].astype(int)
consulta.head()
```

```python
columns = consulta.columns
new_columns = {}
for i in columns:
    new_columns[i] = rem_accents(i).lower().replace(" ","_").replace(".","")
columns = np.delete(columns, [1])
columns = np.delete(columns, [2])
columns
```

```python
columns = consulta.columns
new_columns = {}
for i in columns:
    new_columns[i] = rem_accents(i).lower().replace(" ","_")

for i in columns:
    if i != "id":
        consulta[i] = consulta[i].apply(lambda x: rem_accents(str(x).lower()))

consulta.rename(columns=new_columns, inplace=True)

consulta["fecha_consulta"] = pd.to_datetime(consulta["fecha_consulta"])
consulta["fecha_nacimiento"] = pd.to_datetime(consulta["fecha_nacimiento"])

consulta.head()
```

```python
import nltk
def find_talla(sentence):
    words = re.sub(r"[;:-]|cms|cm", "", sentence.lower())
    words = nltk.word_tokenize(words)
    df = pd.DataFrame(nltk.pos_tag(words),columns=["token","type"])
    index = df[df["token"].isin(["talla"])].index
    if len(index) > 0 and df.loc[index[0]+1,"type"] == "CD":
        try:
            value = float(df.loc[index[0]+1,"token"])
            if value < 3:
                value *= 100
        except:
            value = np.nan
        
        return value
    else:
        return np.nan
    
def find_peso(sentence):
    words = re.sub(r"[;:-]|kg|kgs", "", sentence.lower())
    words = nltk.word_tokenize(words)
    df = pd.DataFrame(nltk.pos_tag(words),columns=["token","type"])
    index = df[df["token"].isin(["peso"])].index
    if len(index) > 0 and df.loc[index[0]+1,"type"] == "CD":
        try:
            value = float(df.loc[index[0]+1,"token"])
        except:
            value = np.nan
        
        return value
    else:
        return np.nan
```

```python
consulta["talla"] = consulta["examen_fisico"].apply(find_talla)
consulta["peso"] = consulta["examen_fisico"].apply(find_peso)
```

```python
consulta["imc"] = round(consulta["peso"] / ((consulta["talla"]/100)**2),1)
consulta.drop(["talla","peso"],inplace=True,axis=1)
```

```python
for name, group in consulta.groupby(["id","fecha_consulta"]):
    index = group.index
    consulta.loc[index, "imc"] = np.nanmean(consulta.loc[index,"imc"])

consulta.drop_duplicates(subset=["id","fecha_consulta","diagnostico_principal"],inplace=True)
```
