```python
import unicodedata
def rem_accents(text):
    return unicodedata.normalize('NFD', text)\
           .encode('ascii', 'ignore')\
           .decode("utf-8")
```

```python
columns = df_Fenotipos_v6.columns
new_columns = {}
for i in columns:
    new_columns[i] = rem_accents(i).lower().replace(" ","_")
```

```python
for i in columns:
    df_Fenotipos_v6[i] = df_Fenotipos_v6[i].apply(lambda x: rem_accents(str(x).lower()))
```

```python
df_Fenotipos_v6.rename(columns=new_columns, inplace=True)
df_Fenotipos_v6.head()
```

```python
df_Fenotipos_v6["fecha_consulta"] = pd.to_datetime(df_Fenotipos_v6["fecha_consulta"])
```

```python
df_Fenotipos_v6['ano']=df_Fenotipos_v6['fecha_consulta'].dt.year
df_Fenotipos_v6['mes']=df_Fenotipos_v6['fecha_consulta'].dt.month
df_Fenotipos_v6['dia']=df_Fenotipos_v6['fecha_consulta'].dt.day
df_Fenotipos_v6
```
