```python
df_Fenotipos_v2 = df_Fenotipos.groupby(["Fenotipo(s)"]).count()
df_Fenotipos_v2.reset_index(inplace=True)
df_Fenotipos_v2.sort_values(by=['ID'],ascending=False)
```

```python
df_Fenotipos_v3 = pd.DataFrame(df_Fenotipos_v2['Fenotipo(s)'].str.split(pat="|").values.tolist())
df_Fenotipos_v3
```

```python
c1=df_Fenotipos_v3[0]
c2=df_Fenotipos_v3[1]
c3=df_Fenotipos_v3[2]
c4=df_Fenotipos_v3[3]
print(len(c1)+len(c2)+len(c3)+len(c4))
```

```python
df_Fenotipos_v4 = c1.append(c2, ignore_index=True)
df_Fenotipos_v4 = df_Fenotipos_v4.append(c3, ignore_index=True)
df_Fenotipos_v4 = df_Fenotipos_v4.append(c4, ignore_index=True)
df_Fenotipos_v4 = df_Fenotipos_v4.str.strip()
df_Fenotipos_v4 = df_Fenotipos_v4.to_frame()
df_Fenotipos_v4
```
